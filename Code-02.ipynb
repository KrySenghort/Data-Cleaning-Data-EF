### Dataset 07: Cambodia's Economic Overview and Future Outlook (2015-2029)

```python
import requests

url = "https://data.mef.gov.kh/api/v1/public-datasets/pd_681bb9e6dbc953000126f5c4/json?page=1&page_size=100"
response = requests.get(url)
data02 = response.json()

# Extract the 'items' list
items = data02['items']
df = pd.DataFrame(items)
df = df.drop(columns=['row_number'])
df.head()
```

```python
df = df.transpose()
import pandas as pd

# Assume df is your DataFrame

# Identify columns that are years (they're numeric strings or integers)
year_columns = [col for col in df.columns if str(col).isdigit()]

# Melt the DataFrame to long format
df_long = pd.melt(
    df,
    id_vars=[col for col in df.columns if col not in year_columns],  # keep text columns
    value_vars=year_columns,                                         # melt only year columns
    var_name='Year',                                                 # new column for years
    value_name='Value'                                               # new column for data values
)

# Convert Year to integer if needed
df_long['Year'] = df_long['Year'].astype(int)

# Display result
df_long.head()
```

```python
# Convert to integer year safely
df_long['estimates start after'] = df_long['estimates start after'].astype(float).astype('Int64')
df_long['Year'] = df_long['Year'].astype(int)
df_long 
```

```python
# Reorder and rename columns for clarity
df_long = df_long[[
    'Year',
    'estimates start after',
    'country',
    'subject descriptor',
    'units',
    'Value',
    'scale',
    'country/series-specific notes'
]]
df_long 
```

```python
df_long.to_csv("07-Cambodia's Economic Overview and Future Outlook (2015-2029).csv", index=False)
```

### Dataset 08: Cambodia's Macroeconomic Outlook: Recovery and Projections (2019–2026)

```python
import requests
import pandas as pd

url = "https://data.mef.gov.kh/api/v1/public-datasets/pd_681c1151dbc953000126f5ca/json?page=1&page_size=10"
response = requests.get(url)
data02 = response.json()

# Extract the 'items' list
items = data02['items']
df = pd.DataFrame(items)
df = df.drop(columns=['row_number'])
df.head()
```

```python
import pandas as pd

# Step 1: Melt the DataFrame to long format
df_melted = df.melt(id_vars=['no', 'sector'], 
                    var_name='year_raw', 
                    value_name='value')

# Step 2: Extract year, is_estimated, is_predicted
df_melted['is_estimated'] = df_melted['year_raw'].str.endswith('e')
df_melted['is_predicted'] = df_melted['year_raw'].str.endswith('p')
df_melted['year'] = df_melted['year_raw'].str.extract(r'(\d{4})').astype(int)

# Step 3: Final normalized structure
df_normalized = df_melted[['no', 'sector', 'year', 'value', 'is_estimated', 'is_predicted']]

# Optional: sort for readability
df_normalized = df_normalized.sort_values(['no', 'year'])
df_normalized
```

#### Dataset 09: Foreign Tourism Statistics and Revenue from Koh Ker Temple of FY2021

```python
import requests
import pandas as pd

all_items = []

# Loop through all 146 pages
for page in range(1, 2):
    url = f"https://data.mef.gov.kh/api/v1/public-datasets/pd_6768c749255e6c0001248280/json?page={page}&page_size=100"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        items = data.get('items', [])
        all_items.extend(items)
    else:
        print(f"Failed to fetch page {page}")

# Create DataFrame from all items
df09 = pd.DataFrame(all_items)

# Drop 'row_number' column if it exists
if 'row_number' in df09.columns:
    df09 = df09.drop(columns=['row_number'])

df09
```

```python
df09 = df09.rename(columns={
    'ទេសចរណ៍ទិញបណ្ណ' : 'Month',
    'ចំនួនភ្ញៀវ' : 'Number of Visitors',
    'ចំណូល(ដុល្លារ)' : 'Revenue(USD)',
})
df09
```

```python
# Khmer to English month mapping
khmer_to_english_months = {
    'មករា': 'January',
    'កុម្ភៈ': 'February',
    'មីនា': 'March',
    'មេសា': 'April',
    'ឧសភា': 'May',
    'មិថុនា': 'June',
    'កក្កដា': 'July',
    'សីហា': 'August',
    'កញ្ញា': 'September',
    'តុលា': 'October',
    'វិច្ឆិកា': 'November',
    'ធ្នូ': 'December'
}

# Replace Khmer month names with English
df09['Month'] = df09['Month'].map(khmer_to_english_months)
df09
```

```python
# Assigned Year = 2021
year = 2021
# Create a date string and convert to datetime format
df09['Month'] = pd.to_datetime(df09['Month'] + '-' + str(year))
df09
```

```python
df09.to_csv('09-Foreign Tourism Statistics and Revenue from Koh Ker Temple of FY2021.csv', index=False)
print("Data saved to '09-Foreign Tourism Statistics and Revenue from Koh Ker Temple of FY2021.csv'")
```
