#### Calling Data of Geography from Json

```python
import requests

url = "https://data.mef.gov.kh/api/v1/public-datasets/pd_66a8603700604c000123e144/json?page=1&page_size=100"
response2 = requests.get(url)
data02 = response2.json()

# Extract the 'items' list
items = data02['items']
df_geo = pd.DataFrame(items)
df_geo.head()
```

```python
df_geo['full_name'] = df_geo['type'] + df_geo['name_km']
df_geo.head(n=10)
```

#### Create one key column of targeted dataset to transform from Khmer to English as a whole

```python
df02['Capital_Province'] = df_geo['name_en'].astype(str).str.replace(' ', ' ', regex=False)
df02
```

```python
df02['Capital_Province'] = df02['Capital_Province'].fillna('National Hospital')
df02
```

```python
df02.to_csv('02-Administering Vitamin A Supplementation - Cambodia.csv', index=False)
```

### Dataset 04: Cambodia Loan signed by sector of public external 2022
(https://data.mef.gov.kh/datasets/pd_66b498c349eb6c00013c129a)

```python
import requests
import pandas as pd 

url = "https://data.mef.gov.kh/api/v1/public-datasets/pd_66b498c349eb6c00013c129a/json?page=1&page_size=10"
response = requests.get(url)
data07 = response.json()

# Extract the 'items' list
items = data07['items']
df = pd.DataFrame(items)
df = df.drop(columns=['row_number'])
df.head()
```

```python
df_transposed = df.transpose()

df_transposed.columns = df_transposed.iloc[0]  # Set first row as header
df_transposed = df_transposed[1:]              # Remove the first row

# Step 3: Reset index (optional, if you don't want row labels)
# df_transposed.reset_index(drop=True, inplace=True)
df_transposed.head()
```

```python
df_transposed.rename(columns={'description': 'Quarter'}, inplace=True)
df_transposed
```

```python
df_transposed
```

```python
df_transposed.to_csv('03-Cambodia Loan signed by sector of public external 2022.csv')
```
